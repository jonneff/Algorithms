```python
"""
ALP 13.1 Group anagrams

Given an array of strings, group anagrams together.

Trick:  Sort characters in string and use as a key in a dictionary.  If two words have the same sorted string, 
they are anagrams.  

Complexity:  O(nm log nm) where n = #strings and m = max length of strings.
"""
def groupAnagrams(self, strs):
    sDict = {}
    for s in strs:
        li = list(s)
        li.sort()
        sort = ''.join(li) 
        if sort in sDict:
            sDict[sort].append(s)
        else:
            sDict[sort] = [s]

# Leetcode requires sorted inner lists
    ret = []
    for key in sDict:
        inner = sDict[key]
        inner.sort()
        ret.append(inner)
         
    return ret
```

```python
"""
Contains Duplicate:  Given an array of integers, find if the array contains any duplicates. Your function should 
return true if 
any value appears at least twice in the array, and it should return false if every element is distinct.
"""
class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        numset = set(nums)
        if len(nums) > len(numset):
            return True
        return False
```

```python
"""
Shortest word distance:  Given a list of words and two words word1 and word2, return the shortest distance between 
these two 
words in the list.  For example,
Assume that words = ["practice", "makes", "perfect", "coding", "makes"]. 
Given word1 = “coding”, word2 = “practice”, return 3.
"""
import sys
class Solution(object):
    def shortestDistance(self, words, word1, word2):
        """
        :type words: List[str]
        :type word1: str
        :type word2: str
        :rtype: int
        """
        i1 = []
        i2 = []
        for i in range(len(words)):
            if words[i] == word1:
                i1.append(i)
            if words[i] == word2:
                i2.append(i)
        
        q = sys.maxint
        for i in range(len(i1)):
            for j in range(len(i2)):
                dist = abs(i1[i] - i2[j])
                if q > dist:
                    q = dist
        return q
```

```python
"""
Move Zeroes:  Given an array nums, write a function to move all 0's to the end of it while maintaining the relative 
order of the 
non-zero elements. 
For example, given nums = [0, 1, 0, 3, 12], after calling your function, nums should be [1, 3, 12, 0, 0]. 
Note:
1.	You must do this in-place without making a copy of the array.
2.	Minimize the total number of operations.
"""
class Solution(object):
    def moveZeroes(self, nums):
        """
        :type nums: List[int]
        :rtype: void Do not return anything, modify nums in-place instead.
        """
        ptr = 0
        nzeroes = 0
        while ptr < len(nums):
            if nums[ptr] == 0:
                del nums[ptr]
                nzeroes+=1
            else:
                ptr+=1
        nums.extend(nzeroes*[0])
```

```python
"""
Meeting Rooms:  Given an array of meeting time intervals consisting of start and end times [[s1,e1],[s2,e2],...] 
(si < ei), 
determine if a person could attend all meetings.
For example, Given [[0, 30],[5, 10],[15, 20]], return false.
"""

# Definition for an interval.
# class Interval(object):
#     def __init__(self, s=0, e=0):
#         self.start = s
#         self.end = e

class Solution(object):
    def canAttendMeetings(self, intervals):
        """
        :type intervals: List[Interval]
        :rtype: bool
        """
        intervals.sort(key = lambda x: x.start)
        for i in range(len(intervals)-1):
            if intervals[i].end > intervals[i+1].start:
                return False
        return True
```

```python
"""
Majority Element:  Given an array of size n, find the majority element. The majority element is the element that 
appears more 
than ⌊ n/2 ⌋ times.  You may assume that the array is non-empty and the majority element always exist in the array.
"""
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        ndict = {}
        for n in nums:
            if n in ndict:
                ndict[n]+=1
            else:
                ndict[n]=1
        for k in ndict:
            if ndict[k] > len(nums)/2:
                return k
```

```python
"""
Wiggle Sort:  Given an unsorted array nums, reorder it in-place such that nums[0] <= nums[1] >= nums[2] <= nums[3]....
For example, given nums = [3, 5, 2, 1, 6, 4], one possible answer is [1, 6, 2, 5, 3, 4].
"""
class Solution(object):
    def wiggleSort(self, nums):
        """
        :type nums: List[int]
        :rtype: void Do not return anything, modify nums in-place instead.
        """
        nums.sort()
        l = len(nums)
        mid = l//2
        a = nums[:mid]
        b = nums[mid:]
        for i in range(mid):
            nums[2*i], nums[2*i+1] = a[i], b[i]
        if l%2 != 0:
            nums[l-1], nums[l-2] = nums[l-2], nums[l-1]
```

```python
"""
Strobogrammatic Number:  A strobogrammatic number is a number that looks the same when rotated 180 degrees (looked at 
upside down).  Write a function to determine if a number is strobogrammatic. The number is represented as a string.  
For example, the numbers "69", "88", and "818" are all strobogrammatic.

Trick:  This is one of many problems where you have a string or list and two pointers working together.  
Frequently, you start at left and right end and work towards the middle.  
"""
class Solution(object):
    def isStrobogrammatic(self, num):
        """
        :type num: str
        :rtype: bool
        """
        l = 0
        r = len(num)-1
        while l < r:
            if num[l] in ['0','1','8'] and num[l] == num[r]:
                l+=1
                r-=1
                continue
            if (num[l]=='6' and num[r]=='9') or (num[l]=='9' and num[r]=='6'):
                l+=1
                r-=1
                continue
            return False
        if l == r:
            if num[l] not in ['0','1','8']:
                return False
        return True
```

```python
"""
Remove Element:  Given an array and a value, remove all instances of that value in place and return the new length.  
The order of elements can be changed. It doesn't matter what you leave beyond the new length.

Note:  list comprehension won’t work because it creates a NEW list, even if you set it equal to original list.  i.e., 
nums = [x for x in nums if x!=val] won’t pass Leetcode. 
"""
class Solution(object):
    def removeElement(self, nums, val):
        """
        :type nums: List[int]
        :type val: int
        :rtype: int
        """
        while val in nums:
            nums.remove(val)
        return len(nums)
```

```python
"""
Remove Duplicates From Sorted Array:  Given a sorted array, remove the duplicates in place such that each element 
appear only once 
and return the new length.  Do not allocate extra space for another array, you must do this in place with constant 
memory.  For 
example, Given input array nums = [1,1,2],
Your function should return length = 2, with the first two elements of nums being 1 and 2 respectively. It doesn't 
matter what you 
leave beyond the new length.

Trick:  Don’t delete anything, just change values and leave garbage at end of list, which Leetcode doesn’t check. 

NOTE:  Fast and slow pointer approach is canonical solution. 
"""
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums) == 0:
            return 0
        # i is slow pointer, j is fast pointer
        i = 0
        j = 1
        while j < len(nums):
            if nums[j] != nums[i]:
                i+=1
                nums[i] = nums[j]
            j+=1
        return i+1
```

```python
"""
Longest Common Prefix My Submissions Question
Total Accepted: 74964 Total Submissions: 283974 Difficulty: Easy
Write a function to find the longest common prefix string amongst an array of strings.
"""
"""
Beats 59%.  

zip(*strs) returns an iterable with tuples.  If strs =['catastrophe','cats'] then
zip(*strs) returns [('c','c'),('a','a'),('t','t'),('a','s')]
By using set() you find out if the corresponding letters in the same position in string are the same.
"""
class Solution(object):
    def longestCommonPrefix(self, strs):
        a = zip(*strs)
        pre = []
        for tup in a:
            if len(set(tup)) !=1:
                break
            pre.append(tup[0])
        return "".join(pre)
```

```python
"""
Shortest Word Distance III My Submissions Question
Total Accepted: 3381 Total Submissions: 7677 Difficulty: Medium
This is a follow up of Shortest Word Distance. The only difference is now word1 could be the same as word2.

Given a list of words and two words word1 and word2, return the shortest distance between these two words in the list.

word1 and word2 may be the same and they represent two individual words in the list.

For example,
Assume that words = ["practice", "makes", "perfect", "coding", "makes"].

Given word1 = “makes”, word2 = “coding”, return 1.
Given word1 = "makes", word2 = "makes", return 3.

Note:
You may assume word1 and word2 are both in the list.
"""
"""
Mine beats 85%
"""
class Solution(object):
    def shortestWordDistance(self, words, word1, word2):
        # word can appear multiple times; have to find every index
        lw = len(words)
        inds = {word1:[], word2:[]}
        for i in range(lw):
            if words[i] == word1:
                inds[word1].append(i)
            if words[i] == word2:
                inds[word2].append(i)
        diff = []
        for ind1 in inds[word1]:
            for ind2 in inds[word2]:
                if ind1 != ind2:
                    diff.append(abs(ind1-ind2))
        return min(diff)
```

```python
"""
Search a 2D Matrix II My Submissions Question
Total Accepted: 21381 Total Submissions: 68413 Difficulty: Medium
Write an efficient algorithm that searches for a value in an m x n matrix. This matrix has the following properties:

Integers in each row are sorted in ascending from left to right.
Integers in each column are sorted in ascending from top to bottom.
For example,

Consider the following matrix:

[
  [1,   4,  7, 11, 15],
  [2,   5,  8, 12, 19],
  [3,   6,  9, 16, 22],
  [10, 13, 14, 17, 24],
  [18, 21, 23, 26, 30]
]
Given target = 5, return true.

Given target = 20, return false.
"""
"""
Seb showed me this solution.  I coded it and it beats 82% of Python submissions.
"""
class Solution(object):
    def searchMatrix(self, matrix, target):
        # start in lower left corner.  
        i, j = len(matrix)-1, 0
        n = len(matrix[0])
        while i >= 0 and j < n:
            if matrix[i][j] > target:
                i -= 1
                continue
            if matrix[i][j] < target:
                j += 1
                continue
            if matrix[i][j] == target:
                return True
        return False
```

```python
"""
Symmetric Tree:  Given a binary tree, check whether it is a mirror of itself (ie, symmetric around its center).
For example, this binary tree is symmetric:
    1    
   / \   
  2   2  
 / \ / \ 
3  4 4  3 
"""
class Solution(object):
    def isSymmetric(self, root):
        if not root:
            return True
        return self.isSym([root])

    def isSym(self, nodes):
        childvals = []
        for node in nodes:
            if node:
                if node.left:
                    childvals.append(node.left.val)
                else:
                    childvals.append(None)
                if node.right:
                    childvals.append(node.right.val)
                else:
                    childvals.append(None)
        l = 0
        r = len(childvals)-1
        while l < r:
            if childvals[l] != childvals[r]:
                return False
            l+=1
            r-=1
        children = []
        for node in nodes:
            if node:
                if node.left:
                    children.append(node.left)
                if node.right:
                    children.append(node.right)
        if not children:
            return True
        return self.isSym(children)
```

```python
"""
Symmetric Tree, another solution:  This is an elegant solution I found on Leetcode discussion board.  
It uses stack, and so is a form of depth-first search.  There are 5 line Python solutions but I don’t understand them.  

In this solution, the appropriate paired “mirror” nodes on the tree are added to the stack as tuples.  
The trick is that each of these paired tuples must have children that are also mirror pairs.  
"""
class Solution(object):
    def isSymmetric(self, root):
        if not root:
            return True
        stack = []
        stack.append((root.left, root.right))

        while stack:
            left, right = stack.pop()
            if not left and not right:
                continue
            elif left and right and left.val == right.val:
                stack.append((left.left, right.right))
                stack.append((left.right, right.left))
            else:
                return False
            return True
```

```python
"""
M 4.1 Balanced Binary Tree
Given a binary tree, determine if it is height-balanced.  For this problem, a height-balanced binary tree is defined 
as a binary 
tree in which the depth of the two subtrees of every node never differ by more than 1. 

Trick:  Define a recursive helper function checkHeight() which returns -1 if tree is unbalanced, returns height 
otherwise.
Complexity:  Runs in O(n) time; visits all nodes once.  Constant memory.
"""
"""
Beats 91% of Python submissions.  Based on textbook "Cracking the Coding Interview".
"""
class Solution(object):
    def isBalanced(self, root):
        return self.checkHeight(root) != -1
    
    def checkHeight(self, root):
        if root == None:  
            return 0
        leftHeight = self.checkHeight(root.left)
        if leftHeight == -1:
            return -1
        rightHeight = self.checkHeight(root.right)
        if rightHeight == -1:
            return -1
        diffHeight = abs(leftHeight - rightHeight)
        if diffHeight > 1:
            return -1
        else:
            return max(leftHeight, rightHeight) + 1
```

```python
"""
M 4.7 Lowest Common Ancestor of a Binary Tree

Design an algorithm and write code to find the first common ancestor of two nodes in a binary tree. Avoid storing 
additional nodes 
in a data structure. NOTE: This is not necessarily a binary search tree.  

Assumptions:
•	A node is allowed to be a descendent of itself. 
•	If root, p, or q are None, return None.

Complexity:  O(n)
"""
def lowestCommonAncestor(self, root, p, q):
        if root == None:  
            return None
        if root == p or root == q:  
            return root
        l = self.lowestCommonAncestor(root.left, p, q)
        r = self.lowestCommonAncestor(root.right, p, q)
        if l and r:  
            return root 
        if not l and not r:  
            return None
        if l:  
            return l
        if r:  
            return r
```

```python
"""
Binary Tree Level Order Traversal:  Given a binary tree, return the top-down level order traversal of its nodes' 
values. 
(ie, from left to right, level by level from root to leaf).  For example:
Given binary tree,
    3    
   / \   
  9  20     
    /  \    
   15   7 

return its level order traversal as: 
[
  [3],
  [9,20],
  [15,7]
]

Approach:  Do top-down traversal and reverse.  Use 4 lists:  traversal, level, nodes and children.  Each element of 
traversal is 
a level. 
"""
class Solution(object):
    def levelOrderBottom(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root:
            return []
        traversal = []
        nodes = [root]
        i = 1
        while nodes:
            level = []
            children = []
            for node in nodes:
                level.append(node.val)
                if node.left:
                    children.append(node.left)
                if node.right:
                    children.append(node.right)
            traversal.append(level)
            nodes = children
        return traversal
```

```python
"""
Closest Binary Search Tree Value

Given a non-empty binary search tree and a target value, find the value in the BST that is closest to the target.

Note:
Given target value is a floating point.
You are guaranteed to have only one unique value in the BST that is closest to the target.
"""
import sys
class Solution(object):
    def closestValue(self, root, target):
        """
        :type root: TreeNode
        :type target: float
        :rtype: int
        """
        # iteration
        node = root
        currClosest = sys.maxint
        while node:
            if node.val == target:
                return node.val
            if abs(node.val - target) < abs(currClosest - target):
                currClosest = node.val
            if target < node.val:
                node = node.left
            else:
                node = node.right
        return currClosest
```

```python
"""
Path Sum 
Given a binary tree and a sum, determine if the tree has a root-to-leaf path such that adding up all the values along 
the path equals the given sum.

For example:
Given the below binary tree and sum = 22,
              5
             / \
            4   8
           /   / \
          11  13  4
         /  \      \
        7    2      1
return true, as there exist a root-to-leaf path 5->4->11->2 which sum is 22.
"""

class Solution(object):
    def hasPathSum(self, root, sum):
        if root == None:
            return False
        if root.left == root.right == None and root.val == sum:
            return True
        sum = sum - root.val
        return self.hasPathSum(root.left, sum) or self.hasPathSum(root.right, sum)
```

```python
"""
Graph Valid Tree My Submissions Question
Total Accepted: 5203 Total Submissions: 18562 Difficulty: Medium
Given n nodes labeled from 0 to n - 1 and a list of undirected edges (each edge is a pair of nodes), write a function 
to check whether these edges make up a valid tree.

For example:

Given n = 5 and edges = [[0, 1], [0, 2], [0, 3], [1, 4]], return true.

Given n = 5 and edges = [[0, 1], [1, 2], [2, 3], [1, 3], [1, 4]], return false.
"""
"""
My 55% solution based on Cormen using DFS.  Using deque and 'w' instead of 0 for "visited" makes it run faster.
"""
class Solution(object):
    def validTree(self, n, edges):
        """
        :type n: int
        :type edges: List[List[int]]
        :rtype: bool
        """
        # a tree is a graph with no cycles
        #  AND ALL NODES MUST BE CONNECTED
        # if you do DFS and encounter a gray node, you have discovered a cycle
        # first create adjacency list from edges
        from collections import deque
        if not edges:
            if n == 1:
                return True
            else:
                return False
        Adj = {i:[] for i in range(n)}
        # edges = map(lambda x: [min(x),max(x)], edges) # have to have min node first for this to work
        for edge in edges:
            Adj[edge[0]].append(edge[1])
            Adj[edge[1]].append(edge[0])
        # create dict to store color for each vertex (need predecessors but not distance)
        V = {i:['w',None] for i in range(n)} # color all nodes white
        S = deque() # stack for DFS
        # initialize with first node in edges (0 is not always in in edge[0] )
        V[edges[0][0]][0] = 'g' # color first node as gray
        S.append(edges[0][0]) # put first node in stack
        while S:
            u = S.pop()
            for v in Adj[u]:
                if V[v][0] == 'g' and V[v][1] != u:
                    return False # if we see a node seen before and it's parent is not u, there's a cycle
                if V[v][0] == 'w':
                    V[v][0] = 'g'
                    V[v][1] = u
                    S.append(v)
            V[u][0] = 'b'
        for node in V:
            if V[node][0] == 'w':
                return False # if any nodes are white, graph is not connected
        return True
```

```python
"""
Course Schedule II My Submissions Question
Total Accepted: 16247 Total Submissions: 79240 Difficulty: Medium
There are a total of n courses you have to take, labeled from 0 to n - 1.

Some courses may have prerequisites, for example to take course 0 you have to first take course 1, which is expressed 
as a pair: [0,1].  Given the total number of courses and a list of prerequisite pairs, return the ordering of courses 
you should take to finish all courses.  There may be multiple correct orders, you just need to return one of them. 
If it is impossible to finish all courses, return an empty array.  For example:

2, [[1,0]]
There are a total of 2 courses to take. To take course 1 you should have finished course 0. So the correct course 
order is [0,1]

4, [[1,0],[2,0],[3,1],[3,2]]
There are a total of 4 courses to take. To take course 3 you should have finished both courses 1 and 2. Both courses 
1 and 2 should be taken after you finished course 0. So one correct course order is [0,1,2,3]. Another correct ordering 
is[0,2,1,3].
"""
"""
This is a topological sort using modified BFS.
"""
class Solution(object):
    def findOrder(self, numCourses, prerequisites):
        # modified BFS 
        from collections import deque
        Adj = {i:[] for i in xrange(numCourses) }
        Deg = {i:0 for i in xrange(numCourses) }
        for edge in prerequisites:
            Adj[edge[1]].append(edge[0])
            Deg[edge[0]] += 1
        Q = deque()
        path = []
        for i in range(numCourses):
            if Deg[i] == 0:
                Q.append(i)
        if not Q:
            return []
        while Q:
            u = Q.popleft()
            path.append(u)
            for v in Adj[u]:
                Deg[v] -= 1
                if Deg[v] == 0:
                    Q.append(v)
        if len(path) == numCourses:
            return path
        return []
```

```python
"""
M 3.2 Min Stack

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Trick:  Use two stacks, one is the main stack and the other stores the mins.  Min stack only changes when you push or 
pop a min 
from the main stack.  

Complexity:  Time complexity is O(1) for push, pop, top, and min.  Memory is O(n).
"""
class MinStack(object):
    def __init__(self):
        self.stack = []
        self.min = []
        
    def push(self, x):
        if not self.min or x <= self.min[-1]:
            self.min.append(x)
        self.stack.append(x)
        
    def pop(self):
        if self.stack:
            if self.stack[-1] == self.min[-1]:  
                self.min.pop() 
            self.stack.pop()
        
    def top(self):
        if self.stack:
            return self.stack[-1]
        
    def getMin(self):
        if self.min:
            return self.min[-1]
```

```python
"""
Implement Queue Using Stacks:  Implement the following operations of a queue using stacks.
•	push(x) -- Push element x to the back of queue.
•	pop() -- Removes the element from in front of queue.
•	peek() -- Get the front element.
•	empty() -- Return whether the queue is empty.
Notes:
•	You must use only standard operations of a stack -- which means only push to top, peek/pop from top, size, and is 
empty 
    operations are valid.
•	Depending on your language, stack may not be supported natively. You may simulate a stack by using a list or deque 
    (double-ended queue), as long as you use only standard operations of a stack.
•	You may assume that all operations are valid (for example, no pop or peek operations will be called on an empty 
queue).
"""
# Trick:  pop() and peek() have to "pour" stack into temp stack to reverse and get "first" element
class Queue(object):
    def __init__(self):
        self.stack = []
        self.temp = []       
    def push(self, x):
        self.stack.append(x)      
    def pop(self):
        while self.stack:
            self.temp.append(self.stack.pop())
        ret = self.temp.pop()
        while self.temp:
            self.stack.append(self.temp.pop())
        return ret      
    def peek(self):
        while self.stack:
            self.temp.append(self.stack.pop())
        ret = self.temp.pop()
        self.stack.append(ret)
        while self.temp:
            self.stack.append(self.temp.pop())
        return ret  
    def empty(self):
        if len(self.stack) == 0:
            return True
        else:
            return False
```

```python
"""
Stack Using Queues:  Implement the following operations of a stack using queues.

push(x) -- Push element x onto stack.
pop() -- Removes the element on top of the stack.
top() -- Get the top element.
empty() -- Return whether the stack is empty.
Notes:
You must use only standard operations of a queue -- which means only push to back, peek/pop from front, size, and is 
empty operations are valid.
Depending on your language, queue may not be supported natively. You may simulate a queue by using a list or deque 
(double-ended queue), as long as you use only standard operations of a queue.
You may assume that all operations are valid (for example, no pop or top operations will be called on an empty stack).
"""

class Stack(object):
    def __init__(self):
        self.q = []
        self.t = []
        
    def push(self, x):
        self.q.append(x)
        
    def help(self):
        while self.q:
            a = self.q.pop(0)
            if not self.q:
                break
            self.t.append(a)
        self.q = self.t
        self.t = []
        return a
            
    def pop(self):
        self.help()
        
    def top(self):
        b = self.help()
        self.q.append(b)
        return b
        
    def empty(self):
        if len(self.q) == 0:
            return True
        return False
```

```python
"""
Plus One:  Given a non-negative number represented as an array of digits, plus one to the number.  
The digits are stored such that the most significant digit is at the head of the list.
"""
class Solution(object):
    def plusOne(self, digits):
        """
        :type digits: List[int]
        :rtype: List[int]
        """
        # stack
        i = 0
        while digits:
            a = digits.pop()
            if a < 9:
                digits.append(a+1)
                digits.extend(i*[0])
                break
            else:
                i+=1
            if not digits:
                digits.append(1)
                digits.extend(i*[0])
                break
        return digits
```

```python
"""
Two Sum III - Data structure design My Submissions Question
Total Accepted: 6024 Total Submissions: 25019 Difficulty: Easy
Design and implement a TwoSum class. It should support the following operations: add and find.

add - Add the number to an internal data structure.
find - Find if there exists any pair of numbers which sum is equal to the value.

For example,
add(1); add(3); add(5);
find(4) -> true
find(7) -> false
"""
"""
Wrote most of this but had to look up on discussion board to complete.  Thanks Stefan Pochman.

Making local copy of dictionary makes Python run faster.

Need second part of "if" in find() to handle case e.g., find(4) and 2 occurs once.  
"""
class TwoSum(object):
    def __init__(self):
        self.dic = {}
        
    def add(self, number):
        if number in self.dic:
            self.dic[number]+= 1
        else:
            self.dic[number] = 1
        
    def find(self, value):
        dic = self.dic
        for key in dic:
            diff = value - key
            if diff in dic and (diff != key or dic[key] > 1):
                return True
        return False
```

```python
"""
Reverse Linked List:  Reverse a singly linked list.
"""
class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        node = head
        last = None
        while not node == None:
            next = node.next
            node.next = last
            last = node
            node = next
        return last
```

```python
"""
Intersection of Two Linked Lists 
Write a program to find the node at which the intersection of two singly linked lists begins.


For example, the following two linked lists:

A:          a1 → a2
                   ↘
                     c1 → c2 → c3
                   ↗            
B:     b1 → b2 → b3
begin to intersect at node c1.
"""
class Solution(object):
    def getIntersectionNode(self, headA, headB):
        # Trick:  lists are same length AFTER intersection
        # so use difference in length to get nodes lined up
        lenA = self.findLength(headA)
        lenB = self.findLength(headB)
        nodeA, nodeB = headA, headB
        if lenA > lenB:
            nodeA = self.startNode(headA, lenA-lenB)
        elif lenB > lenA:
            nodeB = self.startNode(headB, lenB-lenA)
        while nodeA:
            if nodeA == nodeB:
                return nodeA
            nodeA = nodeA.next
            nodeB = nodeB.next
        return None
        
    def findLength(self, head):
        node = head
        len = 0
        while node:
            len+=1
            node = node.next
        return len
    
    def startNode(self, head, diff):
        node = head
        for i in range(diff):
            node = node.next
        return node
```

```python
"""
LRU Cache My Submissions Question
Total Accepted: 57641 Total Submissions: 371860 Difficulty: Hard
Design and implement a data structure for Least Recently Used (LRU) cache. It should support the following operations: 
get and set.

get(key) - Get the value (will always be positive) of the key if the key exists in the cache, otherwise return -1.
set(key, value) - Set or insert the value if the key is not already present. When the cache reached its capacity, it 
should invalidate the least recently used item before inserting a new item.
"""
# Canonical solution:  use doubly linked list plus dictionary for constant time access to sorted data
# Use dummy head and tail to avoid checking None.  head.next is most recent item and tail.pre is least recently used 
# (LRU) item.
class Node(object):
    def __init__(self, key, val):
        self.val = val
        self.key = key
        self.prev = None
        self.next = None
class LRUCache(object):  
    def __init__(self, capacity):
        self.count = 0
        self.cap = capacity
        self.head = Node(None, None)
        self.tail = Node(None, None)
        self.head.next = self.tail
        self.tail.prev = self.head
        self.dic = {}
    def addNode(self, node):
        # always add new node after head
        node.prev = self.head
        node.next = self.head.next
        self.head.next.prev = node
        self.head.next = node       
    def removeNode(self, node):
        prev = node.prev
        next = node.next
        prev.next = next
        next.prev = prev       
    def moveToHead(self, node):
        self.removeNode(node)
        self.addNode(node)      
    def popTail(self):
        res = self.tail.prev
        self.removeNode(res)
        return res      
    def get(self, key):
        node = self.dic.get(key, None)
        if not node:
            return -1
        # move node to head
        self.moveToHead(node)
        return node.val          
    def set(self, key, value):
        node = self.dic.get(key, None)
        if not node:
            new = Node(key, value)
            self.dic[key] = new
            self.addNode(new)
            self.count+=1
            if self.count > self.cap:
                tail = self.popTail()
                del self.dic[tail.key]
                self.count-=1
        else:
            node.val = value
            self.moveToHead(node)
```

```python
# Simple solution using OrderedDict.  Slightly slower than linked list solution.
class LRUCache(object):  
    def __init__(self, capacity):
        self.cap = capacity
        from collections import OrderedDict
        self.dic = OrderedDict()
    def set(self, key, value):
        val = self.dic.get(key, None)
        if not val:
            self.dic[key] = value
            if len(self.dic) > self.cap:
                self.dic.popitem(False) # Use False for FIFO (least recently used)
        else:
            del self.dic[key]
            self.dic[key] = value
    def get(self, key):
        val = self.dic.get(key, None)
        if not val:
            return -1
        # have to delete and insert to set to top of ordered dict
        del self.dic[key]
        self.dic[key] = val
        return val
```

```python
"""
Flip game:  You are playing the following Flip Game with your friend: Given a string that contains only these two 
characters: 
+ and -, you and your friend take turns to flip two consecutive "++" into "--". The game ends when a person can no 
longer make 
a move and therefore the other person will be the winner. 
Write a function to compute all possible states of the string after one valid move. 
For example, given s = "++++", after one move, it may become one of the following states: 
[   "--++",   "+--+",   "++--" ] 
If there is no valid move, return an empty list [].
"""
class Solution(object):
    def generatePossibleNextMoves(self, s):
        """
        :type s: str
        :rtype: List[str]
        """
        ls = len(s)
        if ls == 0 or ls == 1:
            return []
        result = []
        ptr = 1
        while ptr < ls:
            if s[ptr-1] == '+' and s[ptr] == '+':
                pre = s[:ptr-1]
                post = s[ptr+1:]
                result.append(pre+'--'+post)
            ptr+=1
        return result
```

```python
"""
House Robber:  You are a professional robber planning to rob houses along a street. 
Each house has a certain amount of money stashed, the only constraint stopping you from 
robbing each of them is that adjacent houses have security system connected and it will 
automatically contact the police if two adjacent houses were broken into on the same night.  
Given a list of non-negative integers representing the amount of money of each house, determine 
the maximum amount of money you can rob tonight without alerting the police.
"""
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # dynamic programming (memoization)
        currMax = 0
        prevMax = 0
        for n in nums:
            temp = currMax
            currMax = max(currMax, prevMax+n)
            prevMax = temp
        return currMax
```

```python
""" 
Pascal’s Triangle:  Given numRows, generate the first numRows of Pascal's triangle.  For example, given numRows = 5, 
Return

[     [1],     
      1,1],    
    [1,2,1],   
   [1,3,3,1],  
  [1,4,6,4,1] 
]
"""
class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        if numRows == 0:
            return []
        # create a list at each level and append to pascal list, kind of DP
        pascal = [[1]]
        if numRows == 1:
            return pascal
        pascal.append([1,1])
        if numRows == 2:
            return pascal
        for i in range(2, numRows):
            level = (i+1)*[1]
            for j in range(i-1):
                level[j+1] = pascal[i-1][j] + pascal[i-1][j+1]
            pascal.append(level)
            
        return pascal
```

```python
"""
Paint House My Submissions Question
Total Accepted: 3661 Total Submissions: 9013 Difficulty: Medium
There are a row of n houses, each house can be painted with one of the three colors: red, blue or green. The cost of 
painting each house with a certain color is different. You have to paint all the houses such that no two adjacent 
houses have the same color.

The cost of painting each house with a certain color is represented by a n x 3 cost matrix. For example, costs[0][0] 
is the cost of painting house 0 with color red; costs[1][2] is the cost of painting house 1 with color green, and so 
on... Find the minimum cost to paint all houses.

Note:
All costs are positive integers.
"""
class Solution(object):
    def minCost(self, costs):
        # at each current house, total cost of choosing to paint house one color is 
        # current cost plus min(total cost of painting previous house other two colors)
        lc = len(costs)
        if lc == 0:
            return 0
        phcost = costs[0][:] # have to ad [:] to make copy.  Else you get phcost as a pointer to the existing costs[0]
        chcost = 3*[0]
        for i in range(1,lc):
            chcost[0] = costs[i][0] + min(phcost[1], phcost[2])
            chcost[1] = costs[i][1] + min(phcost[0], phcost[2])
            chcost[2] = costs[i][2] + min(phcost[0], phcost[1])
            phcost[:] = chcost[:]
        return min(phcost)
```

```python
"""
Ugly Number:  Write a program to check whether a given number is an ugly number.  Ugly numbers are positive numbers 
whose prime 
factors only include 2, 3, 5. For example, 6, 8 are ugly while 14 is not ugly since it includes another prime factor 
7.  Note 
that 1 is typically treated as an ugly number.
"""
class Solution(object):
    def isUgly(self, num):
        """
        :type num: int
        :rtype: bool
        """
        if num <= 0 or type(num) != int:
            return False
        if num > 1:
            while num%2 == 0:
                num = num//2
            while num%3 == 0:
                num = num//3
            while num%5 == 0:
                num = num//5
        if num == 1:
            return True
        return False
```

```python
"""
Power of Two:  Given an integer, write a function to determine if it is a power of two.
"""
class Solution(object):
    def isPowerOfTwo(self, n):
        """
        :type n: int
        :rtype: bool
        """
        if n <= 0:
            return False
        binstr = bin(n)[2:]
        ones = 0
        for c in binstr:
            if c == '1':
                ones+=1
            if ones > 1:
                return False
        return True
```

```python
"""
Factorial Trailing Zeroes 
Given an integer n, return the number of trailing zeroes in n!.

Note: Your solution should be in logarithmic time complexity.
"""

class Solution(object):
    def trailingZeroes(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n <= 4:
            return 0
        # factorials add a trailing zero every time there is a 5 in the sequence 
        # (because there are a lot of 2's and 5x2=10 adds a trailing zero)
        # note:  when there is 25, there are two 5's, at 125 there are three 5's, etc.
        while n%5 != 0:
            n-=1
        numzeroes = 0
        while n//5 != 0:
            n = n//5
            numzeroes+=n
        return numzeroes
```

```python
"""
Reverse Bits My Submissions Question
Total Accepted: 44027 Total Submissions: 151156 Difficulty: Easy
Reverse bits of a given 32 bits unsigned integer.

For example, given input 43261596 (represented in binary as 00000010100101000001111010011100), return 964176192 
(represented in binary as 00111001011110000010100101000000).
"""
"""
My solution beats 81% of Python submissions
"""
class Solution(object):
    def reverseBits(self, n):
        # Python strings are immutable so convert to list
        #  bin(n) produces a string like '0b1' for n = 1, pad w/ zeroes to get correct answer
        binlist = list(bin(n))[2:]
        pad = 32 - len(binlist)
        binlist.reverse()
        # .append() appends the LIST, not the elements
        binlist.extend(pad*['0'])
        # str(binlist) doesn't work, this is how to convert list to string
        binstr = ''.join(binlist)
        return int(binstr,2)
```

```python
"""
Count and Say My Submissions Question
Total Accepted: 62555 Total Submissions: 234664 Difficulty: Easy
The count-and-say sequence is the sequence of integers beginning as follows:
1, 11, 21, 1211, 111221, ...

1 is read off as "one 1" or 11.
11 is read off as "two 1s" or 21.
21 is read off as "one 2, then one 1" or 1211.
Given an integer n, generate the nth sequence.

Note: The sequence of integers will be represented as a string.
"""
"""
My 55% solution:
"""
class Solution(object):
    def countAndSay(self, n):
        """
        :type n: int
        :rtype: str
        """
        if n <= 0:
            return ''
        say = '1'
        if n == 1:
            return say
        for _ in range(n-1):
            # here we go with slow and fast pointers
            l = r = 0
            lsay = len(say)
            temp = ''
            while l < lsay:
                num = 0
                while r < lsay and say[l] == say[r]:
                    num+=1
                    r+=1
                temp = temp + str(r-l) + say[l]
                l = r
            say = temp
        return say
```

```python
"""
Count Primes My Submissions Question
Total Accepted: 41638 Total Submissions: 190272 Difficulty: Easy
Description:

Count the number of prime numbers less than a non-negative number, n.
"""
"""
Plain old sieve of Eratosthenes powers my 72% solution
"""
class Solution(object):
    def countPrimes(self, n):
        import math
        if n < 3:
            return 0
        # sieve of Eratosthenes:  multiply primes to eliminate composite (non-prime) numbers
        # Finding number of primes LESS THAN n
        A = n*[1]
        for i in range(2,int(math.sqrt(n))+1):
            if A[i] == 1:
                for j in range(i**2,n,i):
                    A[j] = 0
        # subtract entries for 0 and 1, which are not prime
        return sum(A) - 2
```

```python
"""
Product of Array Except Self My Submissions Question
Total Accepted: 26160 Total Submissions: 66246 Difficulty: Medium
Given an array of n integers where n > 1, nums, return an array output such that output[i] is equal to the product of 
all the elements of nums except nums[i].

Solve it without division and in O(n).

For example, given [1,2,3,4], return [24,12,8,6].

Follow up:
Could you solve it with constant space complexity? (Note: The output array does not count as extra space for the 
purpose of space complexity analysis.)
"""
"""
38%
"""
class Solution(object):
    def productExceptSelf(self, nums):
        # Trick:  two lists, start with 1 and recursively multiply by elements
        # One list starts at front of nums and goes forward, one starts at end and moves in reverse
        p, q = 1, 1
        lnums = len(nums)
        fwd, rev, output = lnums*[0], lnums*[0], lnums*[0]
        for i in range(lnums):
            fwd[i] = p
            p = p*nums[i]
            rev[-i-1] = q
            q = q*nums[-i-1]
        for i in range(lnums):
            output[i] = fwd[i]*rev[i]
        return output
```

```python
"""122. Best Time to Buy and Sell Stock II My Submissions Question
Total Accepted: 70625 Total Submissions: 174838 Difficulty: Medium
Say you have an array for which the ith element is the price of a given stock on day i.

Design an algorithm to find the maximum profit. You may complete as many transactions as you like (ie, buy one and 
sell one share of the stock multiple times). However, you may not engage in multiple transactions at the same time 
(ie, you must sell the stock before you buy again).

32%
"""
class Solution(object):
    def maxProfit(self, prices):
        """
        Greedy algo strategy:  always buy low and sell high.  :)  Decide at each step:  buy, sell, or no action.
        Can track state (am I holding stock?) and price paid in 1 variable.  Track profit in one variable also.
        """
        buy = -1 # can't use None.  if buy = 0, then if buy returns None.
        profit = 0
        for i in range(len(prices)-1):
            if prices[i+1] > prices[i] and buy < 0: # price is about to go up.  local min.  buy.
                buy = prices[i]
            if prices[i+1] < prices[i] and buy >= 0: # price is about to go down.  local max.  sell.  
                profit += prices[i] - buy
                buy = -1
        if buy >= 0:
            profit += prices[-1] - buy
        return profit
```

```python
"""
Excel Sheet Column Number:  Given a column title as appear in an Excel sheet, return its corresponding column number.
For example:
    A -> 1     B -> 2     C -> 3     ...     Z -> 26     AA -> 27     AB -> 28 
"""
class Solution(object):
    def titleToNumber(self, s):
        """
        :type s: str
        :rtype: int
        """
        ptr = 0
        col = 0
        ordA = ord('A')
        while ptr < len(s):
            col = col*26 + ord(s[ptr]) - ordA + 1
            ptr+=1
        
        return col
```

```python
"""
Valid Anagram:  Given two strings s and t, write a function to determine if t is an anagram of s.  For example, 
s = "anagram", 
t = "nagaram", return true.
s = "rat", t = "car", return false. 
Note:  You may assume the string contains only lowercase alphabets.
"""
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        # if 2 strings have same sorted list, they are anagrams
        slist = list(s)
        slist.sort()
        tlist = list(t)
        tlist.sort()
        return slist == tlist
```

```python
"""
Roman to Integer:  Given a roman numeral, convert it to an integer.
Input is guaranteed to be within the range from 1 to 3999.
"""
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        rom = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
        lens = len(s)
        num = 0
        ptr1 = 0
        ptr2 = 1
        while ptr1 < lens:
            a = rom[s[ptr1]]
            if ptr2 >= lens:
                num+=a
                return num
            b = rom[s[ptr2]]
            if a >= b:
                num+=a
                ptr1+=1
                ptr2+=1
            else:
                num+=b-a
                ptr1+=2
                ptr2+=2
        return num
```
